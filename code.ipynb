```python
# convert DATE_ASSESSMENT to datetime
df['DATE_ASSESSMENT'] = pd.to_datetime(df['DATE_ASSESSMENT'], errors='coerce')

# list of illness columns
illness_cols = [
    'ARTHRITIS','STROKE','CANCER','COPD','DIAB','DEMENTIA','DEPRESSION',
    'HEARTFAILURE','HYPERTENSION','ESRD','FT_PROBLEM','HEARING',
    'BOWEL_INCONTINENCE','URINE_INCONTINENCE','ANXTY','MNC_DPRSN','SCHZOPRNIA'
]

# define summary function for each person
def summarize_patient(group):
    first_date = group['DATE_ASSESSMENT'].min()
    last_date = group['DATE_ASSESSMENT'].max()
    months_between = (last_date.year - first_date.year) * 12 + (last_date.month - first_date.month)

    summary = {
        'INITIAL_AGE': group.loc[group['DATE_ASSESSMENT'] == first_date, 'AGE'].values[0],
        'GENDER': group['GENDER'].dropna().iloc[0] if group['GENDER'].notna().any() else np.nan,
        'RACE': group['RACE'].dropna().iloc[0] if group['RACE'].notna().any() else np.nan,
        'First_Assessment': first_date,
        'Last_Assessment': last_date,
        'Months_Between': months_between,
        'ADL_SCORE_mean': group['ADL_SCORE'].mean(skipna=True),
        'MOBILITY_SP1_mean': group['MOBILITY_SP1'].mean(skipna=True),
        'COG_MS1_mean': group['COG_MS1'].mean(skipna=True),
        'BIMS': group['BIMS'].dropna().iloc[0] if group['BIMS'].notna().any() else np.nan,
    }

    # illness present if any row has a non-null value (i checked and 
    # each col only contains either NaN or the name of the illness)
    for col in illness_cols:
        illness_present = group[col].notna().any()
        summary[col] = int(illness_present)

    return pd.Series(summary)

# apply group-by logic
summary_df = df.groupby('BENE_ID').apply(summarize_patient).reset_index()

# preview result
summary_df.head()
```

```python
# create univariate analysis summary table for all people

def create_summary_table(df):
    # initialize the summary dataframe
    summary = pd.DataFrame(columns=['Category', 'All_N', 'All_%'])
    
    # total number of residents
    total_residents = len(df)
    
    # helper function to add rows to the summary
    def add_row(category, value_counts):
        n = value_counts.sum()
        percent = (n / total_residents) * 100
        return pd.DataFrame({
            'Category': [category],
            'All_N': [n],
            'All_%': [f"{percent:.1f}%"]
        })
    
    # demographics
    # age groups (need to create bins first)
    age_bins = [0, 64, 74, 84, 94, 120]
    age_labels = ['<65', '65-74', '75-84', '85-94', 'â‰¥95']
    df['age_group'] = pd.cut(df['INITIAL_AGE'], bins=age_bins, labels=age_labels, right=True)
    age_counts = df['age_group'].value_counts()
    for age_group in age_labels:
        summary = pd.concat([summary, add_row(f"Age {age_group}", age_counts.loc[[age_group]])], ignore_index=True)
    
    # gender
    gender_counts = df['GENDER'].value_counts()
    for gender in gender_counts.index:
        summary = pd.concat([summary, add_row(f"Gender {gender}", gender_counts.loc[[gender]])], ignore_index=True)
    
    # race
    race_counts = df['RACE'].value_counts()
    for race in race_counts.index:
        summary = pd.concat([summary, add_row(f"Race {race}", race_counts.loc[[race]])], ignore_index=True)
    
    # ADL impairment (need to categorize first)
    adl_bins = [0, 12, 20, 28]
    adl_labels = ['Mild (0-12)', 'Moderate (13-20)', 'Severe (21-28)']
    df['ADL_category'] = pd.cut(df['ADL_SCORE_mean'], bins=adl_bins, labels=adl_labels)
    adl_counts = df['ADL_category'].value_counts()
    for adl_cat in adl_labels:
        summary = pd.concat([summary, add_row(f"ADL {adl_cat}", adl_counts.loc[[adl_cat]])], ignore_index=True)
    
    # mobility impairment
    mobility_map = {
        0: 'No impairment',
        1: 'Mild impairment',
        2: 'Moderate impairment',
        3: 'Severe impairment'
    }
    df['MOBILITY_category'] = df['MOBILITY_SP1_mean'].round().map(mobility_map)
    mobility_counts = df['MOBILITY_category'].value_counts()
    for mob_cat in mobility_map.values():
        summary = pd.concat([summary, add_row(f"Mobility {mob_cat}", mobility_counts.loc[[mob_cat]])], ignore_index=True)
    
    # cognitive Function (using BIMS column)
    cog_counts = df['BIMS'].value_counts()
    for cog_cat in cog_counts.index:
        summary = pd.concat([summary, add_row(f"Cognitive {cog_cat}", cog_counts.loc[[cog_cat]])], ignore_index=True)
    
    # medical conditions list
    conditions = [
        'ARTHRITIS', 'STROKE', 'CANCER', 'COPD', 'DIAB', 'DEMENTIA', 
        'DEPRESSION', 'HEARTFAILURE', 'HYPERTENSION', 'ESRD',
        'FT_PROBLEM', 'HEARING', 'BOWEL_INCONTINENCE', 
        'URINE_INCONTINENCE', 'ANXTY', 'MNC_DPRSN', 'SCHZOPRNIA'
    ]
    
    for condition in conditions:
        if condition in df.columns:
            condition_counts = df[condition].value_counts()
            if 1 in condition_counts.index:
                summary = pd.concat([summary, add_row(condition, condition_counts.loc[[1]])], ignore_index=True)
    
    # add total row at the top
    total_row = pd.DataFrame({
        'Category': ['All Residents'],
        'All_N': [total_residents],
        'All_%': ['100%']
    })
    
    summary = pd.concat([total_row, summary]).reset_index(drop=True)
    
    return summary


summary_table = create_summary_table(summary_df)
summary_table
```
